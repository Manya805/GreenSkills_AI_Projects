```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
# prompt: how to read dataset from drive small code

df = pd.read_csv('IRIS.csv')
df.head()

```

```python
#generate a heatmap using the above data
import matplotlib.pyplot as plt
import seaborn as sns

# Select only the numerical features for correlation calculation
numerical_features = df.select_dtypes(include=['number'])

# Generate the heatmap with the 'RdBu' colormap (red to blue)
sns.heatmap(numerical_features.corr(), annot=True, cmap ='coolwarm', linewidths=0.5)
plt.show() # Add this line to display the heatmap
```

```python
#generate boxplot
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.show()
```

```python
#generate pair plot
import seaborn as sns
sns.pairplot(df,hue = 'species',palette = "husl",size = 2)
plt.show()
```

```python
# generate violin plot
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.violinplot(data=df)
plt.show()
```
